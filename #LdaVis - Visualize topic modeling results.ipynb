## Load topic modeling elements

```python

import gensim
```

```python
#load ldamodel
from gensim.test.utils import datapath
LdaModel = gensim.models.ldamodel.LdaModel
lda_model = LdaModel.load("crosslingual-visualization//data//tutorial_ldamodel")
```

```python
#load corpus
import pickle
with open('crosslingual-visualization//data//tutorial_corpus.pkl', 'rb') as f:
    corpus = pickle.load(f)
```

```python
#load id2word dictionary
from gensim.corpora import Dictionary
id2word = Dictionary.load("crosslingual-visualization//data//tutorial_id2word")
```
