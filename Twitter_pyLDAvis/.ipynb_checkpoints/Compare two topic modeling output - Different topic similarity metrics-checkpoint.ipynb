```python
class PreparedData(namedtuple('PreparedData', ['topic_coordinates', 'topic_info', 'token_table',
                                               'R', 'lambda_step', 'plot_opts', 'topic_order'])):

    def sorted_terms(self, topic=1, _lambda=1):
        """Retuns a dataframe using _lambda to calculate term relevance of a given topic."""
        tdf = pd.DataFrame(self.topic_info[self.topic_info.Category == 'Topic' + str(topic)])
        if _lambda < 0 or _lambda > 1:
            _lambda = 1
        stdf = tdf.assign(relevance=_lambda * tdf['logprob'] + (1 - _lambda) * tdf['loglift'])
        return stdf.sort_values('relevance', ascending=False)

    def to_dict(self):
        return {'mdsDat': self.topic_coordinates.to_dict(orient='list'),
                'tinfo': self.topic_info.to_dict(orient='list'),
                'token.table': self.token_table.to_dict(orient='list'),
                'R': self.R,
                'lambda.step': self.lambda_step,
                'plot.opts': self.plot_opts,
                'topic.order': self.topic_order}

    def to_json(self):
        return json.dumps(self.to_dict(), cls=NumPyEncoder)
```

```python
#word embedding topic similarity metric
#metrica correcta
def distance_topic_i_j(terms_list_i,terms_list_j):
    total_distances_topic_i = 0.0
    not_found_terms = set()
    for term_i in terms_list_i:
        if term_i in wordembedding:
            dist_for_term_i = []
            for term_j in terms_list_j:
                if term_j in wordembedding:
                    dist_for_term_i.append(wordembedding.wv.distance(term_i,term_j))
                else:
                    not_found_terms.add(term_j)
            total_distances_topic_i+=min(dist_for_term_i)
        else:
            not_found_terms.add(term_i)
    if len(not_found_terms)>0:
        print("Not found", not_found_terms)
    return total_distances_topic_i
```

```python
def generar_matrix_baseline_metric(prepared_data_topic_1, prepared_data_topic_2, relevance_score = 0.6, topn=20):
    relevance_score = 0.6
    topn=20
    matrix = []
    categories_row = []
    i=0
    list_prepared_data = [prepared_data_topic_1,prepared_data_topic_2]
    for topic_model_i in list_prepared_data:
        new_order_topics_i = get_new_order_topic(topic_model_i)
        for topic_id_i in new_order_topics_i:
            row=[]
            categories_row.append(categories[i])
            terms_list_i = topic_model_i.sorted_terms(topic=topic_id_i,_lambda=relevance_score)['Term'][:topn]
            for topic_model_j in list_prepared_data:
                new_order_topics_j = get_new_order_topic(topic_model_j)
                for topic_id_j in new_order_topics_j:
                    terms_list_j = topic_model_j.sorted_terms(topic=topic_id_j,_lambda=relevance_score)['Term'][:topn]
                    row.append(distance_topic_i_j(terms_list_i,terms_list_j))
            matrix.append(row)
        i+=1
    matrix = np.asarray(matrix)
    
    return (matrix, categories_row)
```

# Topic Similarity - Jensen Shannon

```python
from scipy.spatial import distance
import sys
```

```python
def generar_matrix_jensen_shannon(lda_model_gensim_topic_1,lda_model_gensim_topic_2, method='original'):
    
    ######Preparar matrices para jensen shannon
    id2word_topic_1 = {x:y for x,y in lda_model_gensim_topic_1.id2word.items()}
    id2word_topic_2 = {x:y for x,y in lda_model_gensim_topic_2.id2word.items()}
    word2id_topic_1 = {y:x for x,y in id2word_topic_1.items()}
    word2id_topic_2 = {y:x for x,y in id2word_topic_2.items()}
    
    #actualizar diccionario de corpus 1
    nueva_id = len(id2word_topic_1)
    new_probability_array = []
    for term, value in word2id_topic_2.items():
        if term not in word2id_topic_1: #añadir palabra
            word2id_topic_1[term] = nueva_id
            nueva_id+=1
            new_probability_array.append(float("-inf"))

    id2word_topic_1 = {y:x for x,y in word2id_topic_1.items()}
    new_probability_array= np.asarray(new_probability_array) 
    id2word_topic_1 = {y:x for x,y in word2id_topic_1.items()}
    
    #actualizar arreglos de corpus 1
    posterior_probabilities_topic_1 = []
    for topic_id in range(len(lda_model_gensim_topic_1.state.get_Elogbeta())):
        posterior_probabilities_topic_1.append(np.concatenate((lda_model_gensim_topic_1.state.get_Elogbeta()[topic_id], new_probability_array)))

    #actualizar arreglos de corpus 2
    posterior_probabilities_topic_2 = []
    for topic_id in range(len(lda_model_gensim_topic_2.state.get_Elogbeta())):
        posterior_probabilities_topic_2.append(np.full(len(posterior_probabilities_topic_1[0]),float("-inf")))
    for topic_id in range(len(lda_model_gensim_topic_2.state.get_Elogbeta())):    
        set_id = set()
        for key, value in word2id_topic_2.items():
            new_id = word2id_topic_1[key] #esta es la nueva id
            if new_id in set_id:
                print("error!", topic_id, key, value, new_id)

            else:
                set_id.add(new_id)
            probability = lda_model_gensim_topic_2.state.get_Elogbeta()[topic_id][value]
            posterior_probabilities_topic_2[topic_id][new_id] = probability
    #####From this point, it is possible to calculate jensen shannon over posterior_probabilities_topic_1 and posterior_probabilities_topic_2
    #forma original
    if method=='original':
        for i in range(len(posterior_probabilities_topic_1)):
            array = posterior_probabilities_topic_1[i] 
            array[array==float("-inf")] = (sys.float_info.max)*-1
            posterior_probabilities_topic_1[i]  = array
        for i in range(len(posterior_probabilities_topic_2)):
            array = posterior_probabilities_topic_2[i] 
            array[array==float("-inf")] = (sys.float_info.max)*-1
            posterior_probabilities_topic_2[i]  = array
    

    if method=='replace_inf_by_min_value':
        for i in range(len(posterior_probabilities_topic_1)):
            array = posterior_probabilities_topic_1[i] 
            min_value = array[np.isfinite(array)].min()
            print("MIN VALUE", min_value)
            array[array==float("-inf")] = min_value
            posterior_probabilities_topic_1[i]  = array
        for i in range(len(posterior_probabilities_topic_2)):
            array = posterior_probabilities_topic_2[i] 
            min_value = array[np.isfinite(array)].min()
            print("MIN VALUE", min_value)
            array[array==float("-inf")] = min_value
            posterior_probabilities_topic_2[i]  = array

    if method=='without_log':
        for i in range(len(posterior_probabilities_topic_1)):
            array = posterior_probabilities_topic_1[i] 
            posterior_probabilities_topic_1[i]  = np.exp(1)**array

        for i in range(len(posterior_probabilities_topic_2)):
            array = posterior_probabilities_topic_2[i] 
            posterior_probabilities_topic_2[i]  = np.exp(1)**array

        
    ##generate matrix
    posterior_probabilities_lists = posterior_probabilities_topic_1+ posterior_probabilities_topic_2
    matrix = []
    for i in range (len(posterior_probabilities_lists)):
        row = []
        for j in range (len(posterior_probabilities_lists)):
            row.append(distance.jensenshannon(posterior_probabilities_lists[i], posterior_probabilities_lists[j]))
        matrix.append(row)
    matrix = np.asarray(matrix)
    
    
    return matrix
```

```python
import numpy as np

from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)


```

```python
# setting distance_threshold=0 ensures we compute the full tree.
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None,affinity="precomputed", linkage="complete")
np.fill_diagonal(masked_data_clustermap, 0)
model = model.fit(masked_data_clustermap)
fig, ax = plt.subplots(figsize=(20,10))
plt.title('Hierarchical Clustering Dendrogram')
# plot the top three levels of the dendrogram
plot_dendrogram(model, truncate_mode='level', p=9)

plt.xlabel("Number of points in node (or index of point if no parenthesis).")


ax.set_xticklabels([textwrap.fill(categories_topic_names[int(label.get_text())], max_width) for label in ax.xaxis.get_majorticklabels()], fontsize=12, rotation=90)


#plt.savefig('results\\CSCW\\text_for_topicmodeling\\automatic_comparison\\pacific_posts_twitter_instagram_baseline_metric.png', bbox_inches='tight')

plt.show()
```

```python
def calculate_metrics(points, kmin, kmax):
    sil=[]
    for k in range(kmin, kmax+1):
        
        m = AgglomerativeClustering(n_clusters = k, affinity='precomputed', compute_full_tree='auto', linkage='complete').fit(points)
        labels = m.labels_
        sil.append(silhouette_score(points, labels, metric = 'precomputed'))
    return sil
```

```python
sils = calculate_metrics(masked_data_clustermap, 2, len(categories_topic_names)-1)
```

```python
plt.figure(figsize=(10,5))
plt.title("Silhouette on Twitter")
plt.plot(range(2,len(categories_topic_names)),sils)
plt.plot(range(2,len(categories_topic_names)),sils,'or')

plt.xlabel("Num Clusters")
plt.ylabel("Silhouette score")
plt.xticks(range(2,len(categories_topic_names)), rotation=90)
plt.show();
```
